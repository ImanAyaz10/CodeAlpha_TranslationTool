```python
# 1. Install dependencies (Run this in a Colab cell first)
# !pip install gradio groq
!pip install gradio
!pip install groq
import os
import gradio as gr
from groq import Groq


# 2. Setup Groq Client
# Replace "YOUR_GROQ_API_KEY" with your actual key if not using env variables
client = Groq(api_key="tool_key")

def translate_text(text, source_lang, target_lang):
    if not text.strip():
        return "Please enter some text to translate."

    try:
        # Constructing the prompt for the LLM
        system_prompt = f"You are a professional translator. Translate the following text from {source_lang} to {target_lang}. Only provide the translation, no conversational filler."

        chat_completion = client.chat.completions.create(
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": text}
            ],
            model="llama-3.3-70b-versatile",
            temperature=0.3
        )

        return chat_completion.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"

# 3. Create the Gradio Interface
with gr.Blocks(theme=gr.themes.Soft(primary_hue="orange")) as demo:
    gr.Markdown("# ðŸŒ Instant AI Translator")
    gr.Markdown("Powered by Groq & Llama 3.3 for near-instant results.")

    with gr.Row():
        with gr.Column():
            input_text = gr.Textbox(label="Enter Text", placeholder="Type here...", lines=5)
            source_lang = gr.Dropdown(
                choices=["English", "Spanish", "French", "German", "Chinese", "Arabic", "Hindi"],
                value="English",
                label="Source Language"
            )
            target_lang = gr.Dropdown(
                choices=["English", "Spanish", "French", "German", "Chinese", "Arabic", "Hindi"],
                value="Spanish",
                label="Target Language"
            )
            translate_btn = gr.Button("Translate Now", variant="primary")

        with gr.Column():
            output_text = gr.Textbox(label="Translated Text", lines=5, interactive=False)
            copy_btn = gr.Button("ðŸ“‹ Copy to Clipboard")
            # This triggers a JavaScript action for the copy button
            copy_btn.click(fn=None, inputs=output_text, js="(x) => {navigator.clipboard.writeText(x); alert('Copied!');}")

    # Connect the logic
    translate_btn.click(
        fn=translate_text,
        inputs=[input_text, source_lang, target_lang],
        outputs=output_text
    )

# 4. Launch the app
if __name__ == "__main__":
    demo.launch()
```
